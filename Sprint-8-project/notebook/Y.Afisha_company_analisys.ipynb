```python
# Criando a tabela de coorte para taxa de retenção mensal

# Criando coluna com o mês da primeira compra para cada usuário
first_purchase = orders.groupby('Uid')['Buy Ts'].min().dt.to_period('M')
orders['FirstPurchaseMonth'] = orders['Uid'].map(first_purchase)

# Criando uma coluna com o número de meses desde a primeira compra
orders['OrderMonth'] = orders['Buy Ts'].dt.to_period('M')
orders['MonthsSinceFirstPurchase'] = (orders['OrderMonth'] - orders['FirstPurchaseMonth']).apply(lambda x: x.n)

# Criando a tabela de coorte
cohort = orders.groupby(['FirstPurchaseMonth', 'MonthsSinceFirstPurchase'])['Uid'].nunique().unstack(fill_value=0)
cohort = cohort.divide(cohort.iloc[:, 0], axis=0) * 100

# Criando array de anotações com formato de percentual
annot_data = np.empty_like(cohort, dtype=object)
for i in range(cohort.shape[0]):
    for j in range(cohort.shape[1]):
        if cohort.iloc[i, j] > 0:
            annot_data[i, j] = f'{cohort.iloc[i, j]:.0f}%'
        else:
            annot_data[i, j] = ''

# Visualizando a tabela de coorte
plt.figure(figsize=(14, 8))
sns.heatmap(cohort, annot=annot_data, fmt='', cmap='Blues', cbar_kws={'label': 'Taxa de retenção (%)', 'shrink': 0.8}, mask=(cohort == 0), linewidths=1, linecolor='white')
plt.title('Cohort retention rate', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Cohort (mês do primeiro acesso)', fontsize=12, labelpad=14)
plt.xlabel('Número de meses desde o primeiro acesso', fontsize=12, labelpad=14)
plt.gca().set_facecolor('white')
plt.tight_layout()
plt.show()
```

```python
# Pedido por usuário com mês
first_orders_df = orders.groupby('Uid').agg({'Buy Ts': 'min'}).reset_index().rename(columns={'Buy Ts': 'first_purchase'})
first_orders_df['first_purchase_month'] = first_orders_df['first_purchase'].dt.to_period('M')

print('Primeiros pedidos por usuário:')
display(first_orders_df.head())

# Merge com Source ID
user_source = visits[['Uid', 'Source Id']].drop_duplicates()
first_orders_df = pd.merge(first_orders_df, user_source, on='Uid')

print('\nPrimeiros pedidos com Source ID:')
display(first_orders_df.head())

# Receita mensal por usuário
orders['month'] = orders['Buy Ts'].dt.to_period('M')
revenue_df = orders.groupby(['Uid', 'month']).agg({'Revenue': 'sum'}).reset_index()
buyers = first_orders_df.merge(revenue_df, on='Uid')

print('\nReceita por usuário e mês:')
display(buyers.head())

# Preparando dados de custos
costs['month'] = costs['dt'].dt.to_period('M')

# Agregando receita e custos por source e mês
revenue_grouped_df = buyers.groupby(['Source Id', 'month']).agg({'Revenue': 'sum'}).reset_index()
costs_grouped_df = costs.groupby(['source_id', 'month']).agg({'costs': 'sum'}).reset_index().rename(columns={'source_id': 'Source Id'})

# Merge final
report_by_source = pd.merge(
    costs_grouped_df, 
    revenue_grouped_df, 
    left_on=['Source Id', 'month'], 
    right_on=['Source Id', 'month'],
    how='outer'
).fillna(0)

report_by_source = report_by_source.sort_values(['Source Id', 'month'])

print('\n\n=== DATAFRAME FINAL CONSOLIDADO ===')
display(report_by_source.head(20))
print(f'\nShape: {report_by_source.shape}')
print(f'\nColunas: {report_by_source.columns.tolist()}')
```

```python
# Preparando dados para LTV por cohort
# Vou usar buyers dataframe que já tem first_purchase_month e dados de receita

# Calculando idade da cohort (meses desde primeira compra)
buyers['age'] = (buyers['month'] - buyers['first_purchase_month']).apply(lambda x: x.n)

# Calculando LTV por usuário (Revenue por usuário e mês)
ltv_data = buyers[['Uid', 'first_purchase_month', 'age', 'Revenue']].copy()

# Agregando LTV (receita média por usuário, por cohort e age)
ltv_by_cohort = ltv_data.groupby(['first_purchase_month', 'age', 'Uid'])['Revenue'].sum().reset_index()
ltv_by_cohort['ltv'] = ltv_by_cohort['Revenue']

# Criando pivot table com LTV acumulado por cohort
output = ltv_by_cohort.pivot_table(
    index='first_purchase_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
)

# LTV acumulado
ltv_acumulado = output.cumsum(axis=1).round(2).fillna('')

# Visualizar como heatmap
fig, ax = plt.subplots(figsize=(14, 10))
heatmap = sns.heatmap(
    output.cumsum(axis=1).round(2), 
    annot=True, 
    fmt='.2f', 
    cmap='YlOrRd', 
    cbar_kws={'label': 'LTV acumulado ($)'},
    linewidths=1, 
    linecolor='white',
    ax=ax
)

# Ajustando o tamanho da fonte da label do colorbar
heatmap.collections[0].colorbar.ax.yaxis.label.set_size(14)

# Ajustando a rotação do eixo Y para 0 (horizontal) e a barra acompanha a altura total
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

ax.set_title('LTV acumulado por coorte ($)', fontsize=16, fontweight='bold', pad=20)
ax.set_ylabel('Mês da primeira compra', fontsize=14, labelpad=14)
ax.set_xlabel('Idade da coorte (meses)', fontsize=14, labelpad=14)

plt.tight_layout()
plt.show()
```

```python
# Reorganizando os dados para calcular LTV por cohort de forma mais estruturada

# Extraindo a primeira compra por usuário
first_order = orders.groupby('Uid')['Buy Ts'].min().reset_index()
first_order.columns = ['Uid', 'first_order']

# Juntando com o dataframe original para ter a informação de primeira compra em cada linha
df_orders = pd.merge(orders, first_order, on='Uid')

# Preparando datas e períodos para análise
df_orders['first_order'] = df_orders['first_order'].dt.date
df_orders['first_order_month'] = pd.to_datetime(df_orders['first_order']).dt.to_period('M')
df_orders['order_month'] = df_orders['Buy Ts'].dt.to_period('M')

# Calculando a idade da coorte (quantos meses desde a primeira compra)
df_orders['age'] = df_orders['order_month'].astype('int64') - df_orders['first_order_month'].astype('int64')

# Contando o número de usuários únicos por coorte (primeira compra por mês)
cohort_sizes = df_orders.groupby('first_order_month').agg({'Uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# Juntando informação de tamanho da coorte com o dataframe
df_orders = pd.merge(df_orders, cohort_sizes, on='first_order_month')

# Agrupando por coorte e idade para calcular receita total
cohorts = df_orders.groupby(['first_order_month', 'age']).agg({'Revenue': 'sum'}).reset_index()

# Juntando com tamanho da coorte para calcular LTV
report_by_cohort = cohorts.merge(cohort_sizes, on='first_order_month')

# Calculando LTV (Lifetime Value) = Receita total por coorte / número de usuários na coorte
report_by_cohort['ltv'] = report_by_cohort['Revenue'] / report_by_cohort['n_buyers']

# Criando tabela dinâmica com LTV acumulado ao longo do tempo
result = report_by_cohort.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv', 
    aggfunc='mean'
).cumsum(axis=1)

# Criando heatmap com LTV acumulado
print('\n')
fig, ax = plt.subplots(figsize=(16, 10))

heatmap = sns.heatmap(
    result.round(2), 
    annot=True, 
    fmt='.2f', 
    cmap='RdYlGn',  # Verde para maiores valores, vermelho para menores
    cbar_kws={'label': 'LTV acumulado ($)', 'shrink': 1.0},
    linewidths=1, 
    linecolor='white',
    ax=ax,
    vmin=0
)

# Ajustando formatação do gráfico
heatmap.collections[0].colorbar.ax.yaxis.label.set_size(12)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
ax.set_title('LTV acumulado por coorte ao longo do tempo', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_ylabel('Coorte (mês da primeira compra)', fontsize=12, labelpad=14)
ax.set_xlabel('Idade da coorte (meses desde primeira compra)', fontsize=12, labelpad=14)

plt.tight_layout()
plt.show()
```

```python
# Visualização do crescimento de LTV ao longo do tempo

import matplotlib.patches as mpatches

fig, ax1 = plt.subplots(figsize=(14, 7))

months = list(range(len(result.columns)))
ltv_por_mes = []
for col in result.columns:
    ltv_por_mes.append(result[col].dropna().mean())

ax1.plot(months, ltv_por_mes, marker='o', linewidth=3, markersize=10, color='#2E86AB', label='LTV médio')
ax1.fill_between(months, ltv_por_mes, alpha=0.3, color='#2E86AB')
ax1.set_xlabel('Meses desde primeira compra', fontsize=12, labelpad=14)
ax1.set_ylabel('LTV acumulado ($)', fontsize=12, labelpad=14)
ax1.set_title('Como o LTV cresce ao longo do tempo', fontsize=14, fontweight='bold', pad=20)
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, max(ltv_por_mes) * 1.2)

# Adicionar anotações
for i, month in enumerate([0, 3, 6, 11]):
    if month < len(ltv_por_mes):
        ax1.annotate(f'$ {ltv_por_mes[month]:.2f}', 
                    xy=(month, ltv_por_mes[month]),
                    xytext=(0, 10), textcoords='offset points',
                    ha='center', fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
# Calculando o CAC (Customer Acquisition Cost) por origem

# Agrupando custos totais por source_id (nome original da coluna em costs)
costs_by_source = costs.groupby('source_id')['costs'].sum()

# Obtendo usuários únicos convertidos por source (de buyers)
converted_users_by_source = buyers.groupby('Source Id')['Uid'].nunique()

# Convertendo costs_by_source index para int para alinhar com Source Id
costs_by_source.index = costs_by_source.index.astype(int)

# Alinhando os índices para merge
cac_data = pd.DataFrame({
    'costs': costs_by_source,
    'users': converted_users_by_source
})

# Preenchendo zeros para sources sem custos ou usuários
cac_data = cac_data.fillna(0)

# Calculando CAC (Customer Acquisition Cost = custos / usuários)
cac_data['cac'] = cac_data['costs'] / cac_data['users']

print('CAC (Customer Acquisition Cost) por origem:')
print()
print(cac_data)
print()
print('\nResumo de CAC:')
for source, row in cac_data.iterrows():
    if row['users'] > 0:
        print(f'Source {int(source)}: ${row["cac"]:.2f} por cliente adquirido')
```

```python
# Preparando dados para ROMI e ROI por cohort
# Calculando LTV, CAC, Custos e Receita por cohort (mês da primeira compra)

if 'costs_by_source' not in locals():
    costs_by_source = costs.groupby('source_id')['costs'].sum()

# Criando Series de custos indexada por Source ID (necessária para cálculos de cohort)
costs_clean = costs_by_source.copy()

# LTV por cohort: receita total / número de usuários únicos por cohort
ltv_cohort_agg = buyers.groupby('first_purchase_month', as_index=True)[['Revenue', 'Uid']].agg({
    'Revenue': 'sum',
    'Uid': 'nunique'
})
ltv_by_cohort_calc = ltv_cohort_agg['Revenue'] / ltv_cohort_agg['Uid']

# CAC por cohort: custo total / número de usuários únicos que iniciaram compra naquele mês
cac_cohort_sources = buyers.groupby('first_purchase_month')['Source Id'].apply(lambda x: list(x.unique()))
cac_cohort_users = buyers.groupby('first_purchase_month')['Uid'].nunique()
cac_by_cohort_calc = pd.Series(index=cac_cohort_sources.index, dtype=float)
for cohort in cac_cohort_sources.index:
    cohort_cost = costs_clean[costs_clean.index.isin(cac_cohort_sources[cohort])].sum()
    cac_by_cohort_calc[cohort] = cohort_cost / cac_cohort_users[cohort] if cac_cohort_users[cohort] > 0 else 0

# Custos totais por cohort
costs_by_cohort_calc = pd.Series(index=cac_cohort_sources.index, dtype=float)
for cohort in cac_cohort_sources.index:
    costs_by_cohort_calc[cohort] = costs_clean[costs_clean.index.isin(cac_cohort_sources[cohort])].sum()

# Receita total por cohort
revenue_by_cohort = buyers.groupby('first_purchase_month')['Revenue'].sum()

# DataFrame com cohorts únicos
romi_cohort_df = pd.DataFrame({
    'Cohort (mês)': sorted(buyers['first_purchase_month'].unique())
}).reset_index(drop=True)

# Criando DataFrame formatado para exibição de ROMI por cohort
romi_cohort_display = pd.DataFrame({
    'Cohort (mês)': romi_cohort_df['Cohort (mês)'].values,
    'LTV': [f'${ltv_by_cohort_calc.get(c, 0):,.2f}' for c in romi_cohort_df['Cohort (mês)'].values],
    'CAC': [f'${cac_by_cohort_calc.get(c, 0):,.2f}' for c in romi_cohort_df['Cohort (mês)'].values],
    'ROMI': [(ltv_by_cohort_calc.get(c, 0) / cac_by_cohort_calc.get(c, 1)) if cac_by_cohort_calc.get(c, 0) > 0 else 0 
             for c in romi_cohort_df['Cohort (mês)'].values]
})

romi_cohort_display['ROMI'] = romi_cohort_display['ROMI'].apply(lambda x: f'{x:.2f}x' if x > 0 else 'N/A')

print('Dados de cohort preparados para análise de ROMI e ROI')
```

```python
# ROMI por cohort
print('\n' + '='*40)
title = 'ROMI por cohort'
print(title.center(40))
print('='*40 + '\n')
display(romi_cohort_display.reset_index(drop=True))

# ROI por cohort (percentual)
print('\n' + '='*50)
title = 'ROI por cohort (mês da primeira compra)'
print(title.center(50))
print('='*50 + '\n')

# Calculando ROI por cohort: (LTV - CAC) / CAC * 100
roi_by_cohort_calc = ((ltv_by_cohort_calc - cac_by_cohort_calc) / cac_by_cohort_calc * 100).round(2)

# Filtrando apenas linhas válidas (onde cohort está em romi_cohort_df)
roi_cohort_df = pd.DataFrame({
    'Cohort (mês)': romi_cohort_df['Cohort (mês)'].values,
    'Receita Total': [revenue_by_cohort.get(c, 0) for c in romi_cohort_df['Cohort (mês)'].values],
    'Custos Total': [costs_by_cohort_calc.get(c, 0) for c in romi_cohort_df['Cohort (mês)'].values],
    'ROI (%)': [roi_by_cohort_calc.get(c, np.nan) for c in romi_cohort_df['Cohort (mês)'].values]
})

# Formatação para exibição
roi_cohort_display = roi_cohort_df.copy()
roi_cohort_display['Receita Total'] = roi_cohort_display['Receita Total'].apply(lambda x: f'${x:,.2f}')
roi_cohort_display['Custos Total'] = roi_cohort_display['Custos Total'].apply(lambda x: f'${x:,.2f}')
roi_cohort_display['ROI (%)'] = roi_cohort_display['ROI (%)'].apply(lambda x: f'{x:,.2f}%' if pd.notna(x) else 'N/A')

display(roi_cohort_display.reset_index(drop=True))
```

```python
# Criando matriz de LTV acumulado por coorte e idade
ltv_matrix = report_by_cohort.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
).cumsum(axis=1)

# Criando matriz de ROMI (dividindo LTV pelo CAC de cada coorte)
romi_matrix = ltv_matrix.copy()
for month in romi_matrix.index:
    # Obtendo o CAC para cada coorte
    cac_value = cohort_cac.loc[cohort_cac['first_order_month'] == month, 'cac'].values[0]
    # ROMI = LTV / CAC (quanto se retorna para cada $1 investido)
    romi_matrix.loc[month] = romi_matrix.loc[month] / cac_value

# Arredondando matriz de ROMI para duas casas decimais
romi_matrix = romi_matrix.round(2)
romi_matrix = romi_matrix.fillna(0)

print('\nMatriz de ROMI (LTV / CAC) por coorte e idade:')
display(romi_matrix.head(10))
```

```python
# Visualizando a tendência de ROMI ao longo da vida do cliente

# Convertendo matriz de ROMI para dataframe para facilitar análise
romi_detailed = romi_matrix.reset_index().melt(
    id_vars='first_order_month', 
    var_name='age', 
    value_name='romi'
)
romi_detailed['age'] = romi_detailed['age'].astype(int)

# Filtrando apenas ROMIs com CAC válido (onde há dados)
romi_detailed = romi_detailed[romi_detailed['romi'] > 0].copy()

# Calculando ROMI médio por mês de vida do cliente (age)
romi_by_age = romi_detailed[romi_detailed['romi'] > 0].groupby('age')['romi'].mean()

# Criando visualização
fig, ax = plt.subplots(figsize=(14, 7))

# Linha de tendência de ROMI
ax.plot(romi_by_age.index, romi_by_age.values, marker='o', linewidth=3, 
        alpha=0.8, markersize=10, color='#1F77B4', label='ROMI médio')

# Linha de break-even (ROMI = 1x)
ax.axhline(y=1, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Break-even (ROMI = 1x)')

# Preenchimento: ganho em verde, prejuízo em vermelho
ax.fill_between(romi_by_age.index, romi_by_age.values, 1, 
                where=(romi_by_age.values >= 1), alpha=0.25, color='green', label='Ganho (ROMI > 1)')
ax.fill_between(romi_by_age.index, romi_by_age.values, 1, 
                where=(romi_by_age.values < 1), alpha=0.25, color='red', label='Prejuízo (ROMI < 1)')

# Anotações nos pontos principais
for x, y in zip(romi_by_age.index, romi_by_age.values):
    ax.annotate(f'{y:.2f}x', xy=(x, y), xytext=(0, 10), 
                textcoords='offset points', ha='center', fontweight='bold', fontsize=10,
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.6))

# Configuração dos eixos e labels
ax.set_xlabel('Meses desde a primeira compra', fontsize=12, labelpad=14)
ax.set_ylabel('ROMI (LTV / CAC)', fontsize=12, labelpad=14)
ax.set_title('Evolução do ROMI ao longo da vida do cliente', fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(range(0, int(romi_by_age.index.max()) + 1, 1))
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='lower right')
ax.set_ylim(0, max(romi_by_age.values) * 1.15)

plt.tight_layout()
plt.show()
```

### Conclusões sobre ROMI

ROMI (Return on Marketing Investment) = LTV / CAC
- ROMI **< 1** = Prejuízo (paga-se mais para adquirir do que se recebe na vida útil)
- ROMI **> 1** = Lucro (retorna mais que o investimento)
- ROMI **1x** = Ponto de equilíbrio (break-even)

- ROMI inicial (mês 0): 0.54x
   - Significa que cada $1 gasto em aquisição retorna `$0.54` na primeira compra
   -   O cliente não paga nem sua própria aquisição na primeira compra

- **Crescimento gradual até break-even:**
  - Mês 0-6: ROMI está entre 0.54x e 0.90x (prejuízo)
  - Mês 7: ROMI = 0.97x (quase break-even)
  - Mês 8: ROMI = 1.05x **aqui começa a ser lucrativo**
  - Mês 11: ROMI = 1.33x (melhor resultado)

- **Insight crítico:**
  - Leva **8 meses** para o cliente se tornar lucrativo
  - Isso significa que uma retenção de menos de 8 meses = prejuízo
  - O pure LTV mascarava essa realidade ao não considerar 
- Todo esforço de retenção que mantém cliente por 8+ meses é justificável (break-even/lucro)
   - Estratégias para elevar retenção são ROI-positivas


**Impacto na Compreensão de Negócio**

| Métrica | Mês 0 | Mês 11 |
|---------|-------|--------|
| **LTV** | $ 4.80 | $ 11.88 (+147%) |
| **CAC** | ~$ 8.90 | ~$ 8.90 (variável) |
| **ROMI** | 0.54x | 1.33x |
| **Status** | ❌ Prejuízo | ✅ Lucro |


### Interpretações:

**CAC supera receita inicial:**
   - O custo para adquirir um cliente é **1.85x maior** que sua primeira compra
   - Isso sugere potencial overinvestimento em marketing inicial (ou ticket-médio muito baixo)
   - Revisar estratégia de aquisição: talvez focar em clientes com maior potencial de repeat purchase

**Qualidade de coorte:**
   - Sugere deterioração na qualidade de clientes recentes ou mudança na estratégia de aquisição

**Problema estrutural: modelo de negócio desafiador**
   - Com CAC `~$ 8.90` e ticket médio `~$ 5-10`, é difícil ser lucrativo rápido
   - Necessário focar em retenção de longo prazo (8+ meses) para viabilizar negócio
   - Alternativa: aumentar ticket médio ou reduzir CAC
